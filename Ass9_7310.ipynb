```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
data = sns.load_dataset('titanic')
data.head()
```

```python
data.describe()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data['age'] = data['age'].fillna(np.mean(data['age']))
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(
x='sex',
y='age',
hue='survived',
data=data,
palette='Set2'
)
plt.title('Plot for Distribution of Age with Respect to Gender and Survival')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.legend(title='Survived (0 = No, 1 = Yes)')
plt.show()
```
