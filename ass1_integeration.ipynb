```python
import pandas as pd
```

```python
cars1 = pd.DataFrame({
    'car_id':[1,2,3,4,5],
    'brand':['Toyota','Honda','Ford','BMW','Audi'],
    'fuel_type':['Petrol','Diesel','Petrol','Diesel','Petrol']
})
```

```python
cars2 = pd.DataFrame({
    'car_id':[1,2,3,4,5],
    'model':['Corolla','City','Mustang','X5','A4'],
    'fuel_type':['Petrol','Petrol','Petrol','Diesel','Petrol']
})
```

```python
print(cars1)
```

```python
print(cars2)
```

```python
pd.merge(cars1,cars2, on='car_id')
```

```python
pd.merge(cars1,cars2, on=['fuel_type', 'car_id'])
```

```python
pd.merge(cars1,cars2, on='fuel_type', how='left')
```

```python
pd.merge(cars1,cars2, on='fuel_type', how='right')
```

```python
pd.merge(cars1,cars2, on='fuel_type', how='inner')
```

```python
pd.merge(cars1,cars2, on='fuel_type', how='outer')
```

```python
pd.concat([cars1,cars2])
```

```python
pd.concat([cars1,cars2], keys=['Dataset1','Dataset2'])
```

```python
pd.concat([cars1,cars2], ignore_index=True)
```

```python
pd.concat([cars1,cars2], axis=1)
```

```python
car_prices = pd.DataFrame(
    [[800000, 20, 5],
     [600000, 18, 4],
     [1200000, 10, 6],
     [1500000, 12, 8],
     [700000, 22, 5]],
    columns=['Price', 'Mileage', 'Age'],
    index=['Car1', 'Car2', 'Car3', 'Car4', 'Car5']
)
```

```python
print(car_prices)
```

```python
print("\n--------Mean--------\n")
print(car_prices.mean())
```

```python
print("\n-------Median-------\n")
print(car_prices.median())
```

```python
print("\n-----------Mode----------\n")
print(car_prices.mode())
```
