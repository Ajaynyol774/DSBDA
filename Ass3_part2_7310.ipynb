```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("Iris.csv")
data.head()
```

```python
data.info()
```

```python
species_list = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']

for Species in species_list:
    print("="*50)
    print(f"Statistical Details for {Species}")
    print("="*50)
    
    pecies_data = data[data['Species'] == Species]
    
    print("\nMean:")
    print(species_data.mean(numeric_only=True))
    
    print("\nStandard Deviation:")
    print(species_data.std(numeric_only=True))
    
    print("\nPercentiles (25%, 50%, 75%):")
    print(species_data.quantile([0.25, 0.50, 0.75], numeric_only=True))
```
